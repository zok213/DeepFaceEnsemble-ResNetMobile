```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:\FPT_Material\Sem 4\CPV301\Source for PE\Image\image1.png')  # Load your image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
# Display the original image
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.title('Original Image')
plt.show()
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)  # Convert RGB to Grayscale

# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axes
plt.title('Grayscale Image')
plt.show()
```

```python
# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)  # Convert RGB to HSV

# Display the HSV image
plt.imshow(hsv_img)
plt.axis('off')  # Hide axes
plt.title('HSV Image')
plt.show()
```

```python
# Coverting the image to YcbCr color space
ycbcr_img = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)  # Convert RGB to YCbCr

# Display the YCbCr image
plt.imshow(ycbcr_img)
plt.axis('off')  # Hide axes
plt.title('YCbCr Image')
plt.show()
```

```python
# Show 4 images in a single figure
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Display each image in the respective subplot

# Original image
axs[0, 0].imshow(img)
axs[0, 0].axis('off')
axs[0, 0].set_title('Original Image')

# Grayscale image
axs[0, 1].imshow(gray_img, cmap='gray')
axs[0, 1].axis('off')
axs[0, 1].set_title('Grayscale Image')

# HSV image
axs[1, 0].imshow(hsv_img)
axs[1, 0].axis('off')
axs[1, 0].set_title('HSV Image')

# YCbCr image
axs[1, 1].imshow(ycbcr_img)
axs[1, 1].axis('off')
axs[1, 1].set_title('YCbCr Image')

plt.tight_layout()
plt.show()
```
